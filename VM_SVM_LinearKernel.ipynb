```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel("data2.xlsx")
data.head()
```

```python
data['Classes'].value_counts()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split
```

```python
bow_transformer = CountVectorizer().fit(data['Sentences'])
bow_transformer.vocabulary_
title_bow = bow_transformer.transform(data['Sentences'])
print(title_bow)
```

```python
dataframe = pd.DataFrame(title_bow.toarray(),columns = bow_transformer.get_feature_names_out())
```

```python
bow_transformer.vocabulary_
```

```python
tokens = bow_transformer.get_feature_names_out()
print(tokens)
```

```python
X = title_bow.toarray()
print(X)
X.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, data['Classes'], test_size= 0.3, random_state= 50)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn import svm
classifier = svm.LinearSVC(C= 1)
classifier.fit(x_train,y_train)
```

```python
prediction = classifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score
cm = confusion_matrix(y_test, prediction)
acc = accuracy_score(y_test, prediction)
f1 = f1_score(y_test, prediction, average='weighted')
print(cm)
print(acc)
print(f1)
```
